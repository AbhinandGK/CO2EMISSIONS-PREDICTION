## Multiple Linear Regression

```python
df_ff=df.drop(['MODELYEAR','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB_MPG','MAKE','MODEL','VEHICLECLASS','FUELTYPE','TRANSMISSION','FUELCONSUMPTION_CITY'],axis=1)
df_ff.head()
```

```python
Y_=df_ff.CO2EMISSIONS
X_=df_ff.drop(['CO2EMISSIONS'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_, Y_, test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
selection = ExtraTreesRegressor()
selection.fit(X_train,y_train)
```

```python
plt.figure(figsize = (12,8))
feat_importances = pd.Series(selection.feature_importances_, index=X_.columns)
feat_importances.nlargest().plot(kind='barh')
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

```python
regr.fit (X_train, y_train)
```

```python
regr.score(X_train, y_train)
```

```python
regr.score(X_test, y_test)
```

## RandomForest Regressor

```python
MAKE=df[['MAKE']]
MAKE=pd.get_dummies(MAKE,drop_first=True)
MAKE.head()
```

```python
MODEL=df[['MODEL']]
MODEL=pd.get_dummies(MODEL,drop_first=True)
MODEL.head()
```

```python
VEHICLECLASS=df[['VEHICLECLASS']]
VEHICLECLASS=pd.get_dummies(VEHICLECLASS,drop_first=True)
VEHICLECLASS.head()
```

```python
FUELTYPE=df[['FUELTYPE']]
FUELTYPE=pd.get_dummies(FUELTYPE,drop_first=True)
FUELTYPE.head()
```

```python
TRANSMISSION=df[['TRANSMISSION']]
TRANSMISSION=pd.get_dummies(TRANSMISSION,drop_first=True)
TRANSMISSION.head()
```

```python
df_f=pd.concat([df,MAKE,MODEL,VEHICLECLASS,TRANSMISSION,FUELTYPE],axis=1)
df_f.head()
```

```python
df_f.drop(['MODELYEAR','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB_MPG','MAKE','MODEL','VEHICLECLASS','FUELTYPE','TRANSMISSION','FUELCONSUMPTION_CITY'],inplace=True,axis=1)
df_f.head()
```

```python
df_f.info()
```

```python
Y=df_f.CO2EMISSIONS
X=df_f.drop(['CO2EMISSIONS'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
selection = ExtraTreesRegressor()
selection.fit(X_train,y_train)
```

```python
plt.figure(figsize = (12,8))
feat_importances = pd.Series(selection.feature_importances_, index=X.columns)
feat_importances.nlargest(25).plot(kind='barh')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
reg_rf = RandomForestRegressor()
reg_rf.fit(X_train, y_train)
```

```python
reg_rf.score(X_train, y_train)
```

```python
reg_rf.score(X_test, y_test)
```
